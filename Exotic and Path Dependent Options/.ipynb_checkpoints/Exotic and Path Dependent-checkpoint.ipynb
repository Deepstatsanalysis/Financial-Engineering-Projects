```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
```

## P1: Fixed Strike Lookback Call and Put
This time we are going to price a 12-month **Fixed Strike Lookback Call and Put** options. WIth interest rate 3\% per annum, current stock price \$98 and strike price \$100. Monte Carlo Simulation used to estimate the price of Call and Put options for the volatilities ranging from 12% to 48%. 

The payoff of the Call option is $(S_{max} - X)^+$, where $S_{max} = {max}\{S_t : t \in [0,T]\}$. And the payoff of the Put option is $(X - S_{min})^+$, where $S_{min} = min\{S_t : t \in [0,T]\}$

```python
S1_0, K1 = 98., 100.
T1, r1 = 1., 0.03
N1, M1 = 50, 50000
sigma1 = np.linspace(0.12, 0.48, 10)

def FixedStrikeLookBackMC(S0, K, T, r, sigma, N, M, opt):
    """Monte Carlo Simulation of Fixed Strike Lookback Call and Put Options
    Args:
        S0 (float): current stock price
        K (float): strike price
        T (float): The term of the time
        r (float): annual interest rate
        sigma (float): volatility
        N (int): time grid
        M (int): simulation paths
    Returns:
        price (float): value of Fixed Strike LookBack Option at t = 0
    """
    dt = T / float(N) # time step
    
    # running Monte Carlo Simulation
    R1 = np.exp((r - 0.5*sigma**2)*dt + \
                sigma*np.sqrt(dt)*np.random.randn(N, M)) # product matrix
    R2 = np.exp((r - 0.5*sigma**2)*dt - \
                sigma*np.sqrt(dt)*np.random.randn(N, M)) # antithetic matrix
    R = np.concatenate((R1, R2), axis = 1)
    
    S = np.cumprod(np.concatenate((S0 * np.ones((1,2*M)), R), axis = 0),\
                   axis = 0) # simulated stock price
    
    if opt == 'C':
        h = np.amax(S, axis = 0)
        payoff = np.maximum(h-K, 0)
    elif opt == 'P':
        h = np.amin(S, axis = 0)
        payoff = np.maximum(K-h, 0)
    else:
        print('Please specify Option Type')
        
    price = np.mean(payoff) * np.exp(-r*T)
    return price

callPrices = [FixedStrikeLookBackMC(S1_0, K1, T1, r1, sigma1_, N1, M1, 'C') for sigma1_ in sigma1]
putPrices = [FixedStrikeLookBackMC(S1_0, K1, T1, r1, sigma1_, N1, M1, 'P') for sigma1_ in sigma1]
plt.figure(figsize = (12, 6))
plt.subplot(121)
plt.plot(sigma1, callPrices)
plt.grid()
plt.title('Fixed Strike Lookback Call Option')
plt.xlabel('Volatility')
plt.ylabel('Price [$]')

plt.subplot(122)
plt.plot(sigma1, putPrices)
plt.grid()
plt.title('Fixed Strike Lookback Put Option')
plt.xlabel('Volatility')
plt.ylabel('Price [$]')

plt.show()
```

## P2: Collateral Loans
We assume the value of a collateral follows a jump-diffusion process:

$$
\frac{dV_t}{V_t^-} = \mu dt + \sigma dW_t + \gamma dJ_t
$$

where $\mu, \sigma, \gamma < 0$ and $V_0$ are fiven, $J$ is a Poisson process with intensity $\lambda_1$, independent of the Brownian Motion process W.

Consider the collateralized loan, with a contract rate per period $r$ and maturity $T$ on the above collateral, and assume the outstanding balance of that loan follows this process:

$$
L_t = a - bc^{12t}
$$

where $a>0, b>0, c>1$ and $L_0$ are given. The boundary $L_T = 0$.

Define the following stopping time:

$$
Q = min \{ t \geq 0:V_t \leq q_t L_t \}
$$

This stopping time is the first time when the relative value of the collateral (with respect to the outstanding loan balance) crosses a threshold which will be viewed as the "optimal exercise boundary" of the option to default.

Define another stopping time, which is the first time an adverse event occurs:

$$
S = min \{t\geq 0: N_t >0\}
$$

Assume that $N_t$ is a Poisson process with intensit of $\lambda_2$.

Define $\tau = min \{Q, S\}$, we assume the embedded default option will be exercised at time $\tau$, if and only if $\tau < T$.
* If the option is exercised at time $Q$ then the payoff to the borrower is $(L_Q - \varepsilon V_Q)^+$
* If the option is exercised at time $S$ then the payoff to the borrower is $|L_S - \varepsilon V_S|$.

```python
def first_nonzero(arr, axis, invalid_val):
    mask = arr!=0
    return np.where(mask.any(axis=axis), mask.argmax(axis=axis), invalid_val)

def Proj6_2func(lambda1, lambda2, T):
    """Value of a collateral following a jump-diffusion process
    Args:
        lambda1 (float): intensity for the Jump Poisson process
        lambda2 (float): intensity for a Poisson process
        T (float): The term of the time
    Returns:
        price (float): value of collateral
        prob (float): default probability
        eT (float): expected exercise time
    """
    # parameter setup
    V0, L0 = 20000, 22000
    mu, sigma, gamma = -0.1, 0.2, -0.4
    r0, delta, alpha, epsilon = 0.02, 0.25, 0.7, 0.95
    beta = (epsilon - alpha) / T
    R = r0 + delta * lambda2
    r = R / float(12)
    n = T * 12
    PMT = L0*r/ (1 - (1+r)**(-n))
    a = PMT / r
    b = PMT / (r * (1+r)**n)
    c = 1 + r
    N_T = 500
    N_MC = 50000
    
    (t, dt) = np.linspace(0, T, num = N_T, retstep = True)

    L = a - b*np.power(c, 12*t)

    qt = alpha + beta * t
    Z = np.random.standard_normal(size = (N_T, N_MC))

    V = np.zeros((N_T, N_MC))
    V = np.concatenate((np.ones((1, N_MC))*V0, V), axis = 0) # value of the loan

    Q = np.zeros((N_T, N_MC))
    S = np.zeros((N_T, N_MC))
    
    for i in range(N_T):
        Q[i, :] = (qt[i] * L[i] - V[i, :]) >= 0
        dJ = np.random.poisson(lambda1 * dt, N_MC) # jump happens or not
        dN = np.random.poisson(lambda2 * dt, N_MC) # incremental counting
        S[i, :] = dN > 0
        V[i+1, :] = V[i, :]*(1 + mu*dt + sigma*np.sqrt(dt)*Z[i, :] + \
                             gamma*dJ) # value update

    Qt = first_nonzero(Q, 0, N_T+1)
    St = first_nonzero(S, 0, N_T+1)
    N_tau = np.minimum(Qt, St)
    ExType = Qt > St

    payoff = np.zeros(N_MC)
    for j in range(N_MC):
        if N_tau[j] < N_T:
            if ExType[j]:
                payoff[j] = np.abs(L[N_tau[j]] - epsilon*V[N_tau[j], j]) *\
                np.exp(-r0*N_tau[j]*dt) #discounted payoff
            else:
                payoff[j] = np.maximum(L[N_tau[j]] - epsilon*V[N_tau[j], j], 0) *\
                np.exp(-r0*N_tau[j]*dt) #discounted payoff
        else:
            payoff[j] = 0 #no excercise

    price = np.mean(payoff)
    prob = np.sum(N_tau < N_T)/len(N_tau)
    eT = np.mean(N_tau[N_tau < N_T])*dt
    return (price, prob, eT)
```

```python
v1, v2, v3 = Proj6_2func(.2,.4,5.)
print('For the required input, the value of default option D is ${v1:f}'.format(v1 = v1))
print('The default Prob is {v2:f}'.format(v2 = v2))
print('The expected exercise time Et is {v3:f} years.'.format(v3 = v3))
```

```python
lambda1 = np.linspace(0.05, 0.4, 8)
lambda2 = np.linspace(0., 0.8, 9)
T = np.linspace(3, 8, 6)

values1 = []
probs1 = []
eTs1 = []

for lam2_ in lambda2:
    for T_ in T:
        v, p ,e = Proj6_2func(0.2, lam2_, T_)
        values1.append(v)
        probs1.append(p)
        eTs1.append(e)

values2 = []
probs2 = []
eTs2 = []

for lam1_ in lambda1:
    for T_ in T:
        v, p ,e = Proj6_2func(lam1_, 0.4, T_)
        values2.append(v)
        probs2.append(p)
        eTs2.append(e)
```
